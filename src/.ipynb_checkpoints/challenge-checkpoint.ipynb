### 1.- Las top 10 fechas donde hay más tweets. Mencionar el usuario (username) que más publicaciones tiene por cada uno de esos días. Debe incluir las siguientes funciones:

#####  RESOLUCION:
Como solucion general a este item, se carga a un dataframe de spark , solo la informacion necesaria, de los tweets contenidos en el arhivo .json.
Luego de esto procedemos a agrupar esta informacion por fecha (año/mes/dia) y por usuario, para contar los tweet por cada uno de ellos, y esto se lo deja en un campo "cantidad_tw_usuario".
Con esta informacion se procede realizar una "windowed function" en donde se ordena estos "cantidad_tw_usuario" de mayor a menor y se le asigna un ranking, para al final solo final quedarme solo con un registro por cada fecha y usuario, el cual contiene el q mas cantidad de tweets ha realizado.
Como dato final se transforma el dataframe generado a una lista de tuplas que es lo que pide la funcion que se retorne.


##### ANALISIS DE MEMORIA

```python
#import necesario para la llamada a la funcion del archivo .py
import q1_memory as q1m
#llamada a la funcion y retorno de la lista de tuplas resolviendo lo solicitado
q1_memory_return = q1m.q1_memory(file_path)
#presentacion de la informacion por pantalla
print(q1_memory_return)
```

El codigo para poder medir la memoria, se encuentra en el archivo "q1_memory.py" en el cual se debe importar la libreria "from memory_profiler import profile", y en cada funcion a ser analizada, se de colocar "@profile".

Segun lo que podemos interpretar en el uso de la memoria, podemos concluir lo siguiente:
- Se ha ejecutado y analizado linea por linea del codigo y su uso respectivo de memoria.
- en cada linea de codigo se puede observar que no hay un incremento para el uso de la memoria.
- la memoria comienza con un valor, el cual se mantiene durante todo el proceso.

##### ANALISIS DE TIEMPO DE EJECUCION

```python
#import necesario para la llamada a la funcion del archivo .py
import q1_time as q1t
import cProfile, pstats, io
from pstats import SortKey

pr = cProfile.Profile()
pr.enable()

#llamada a la funcion y retorno de la lista de tuplas resolviendo lo solicitado
q1_time_return = q1t.q1_time(file_path)
#presentacion de la informacion por pantalla
print(q1_time_return)

pr.disable()
s = io.StringIO()
sortby = SortKey.CUMULATIVE
ps = pstats.Stats(pr, stream=s).sort_stats(sortby)
ps.print_stats()
print(s.getvalue())
```

El codigo para poder medir el tiempo de ejecucion, se lo puede observar en los pasos previos al llamado de la funcion "q1_time(file_path)" para lo cual se deben importar las librerias "cProfile, pstats, io".

Segun lo que podemos interpretar a partir de las estadisticas generadas de tiempos de ejecucion, podemos concluir lo siguiente:
- La funcion se ejecuta con un tiempo total de 1.187 segundos, lo cual es un buen tiempo.
- Tiene un 96% de llamadas a funciones primitivas.
- La mayoria del tiempo empleado en la funcion es en el llamado a la libreria "py4j" que permite la comunicación bidireccional entre Python y Java.
